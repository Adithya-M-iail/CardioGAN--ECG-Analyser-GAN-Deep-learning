```python
import os
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import ecg_plot
```

```python
def generate_ecg_signals(num_samples, output_dir, start_id=0):
    # Generate synthetic ECG signals/images using deepfakeecg.generate
    deepfakeecg.generate(num_samples, output_dir, start_id=start_id, run_device="cpu")
```

```python
# Define the directory path
output_directory = r"C:\Users\ksk_k\Desktop\CARDIOGAN\CardioGAN--ECG-Analyser-GAN-Deep-learning\ecg_plot_images"

# Create the directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)
```

```python
# Desired number of ECG signals
num_signals = 15  # Replace with the desired number

generate_ecg_signals(num_signals, output_directory, start_id=15)
```

```python
# Create the directory if it doesn't exist
os.makedirs("ecg_plots_images", exist_ok=True)
```

```python
import os

# Create the directory if it doesn't exist
os.makedirs("ecg_plots_images", exist_ok=True)

j = 15
for i, filename in enumerate(os.listdir(output_directory), start=15):
    # Load the ECG signal
    ecg = np.loadtxt(os.path.join(output_directory, filename))

    # Prepare the data for visualization
    ecg[:, 2] = ecg[:, 1] - ecg[:, 0]
    ecg[:, 3] = -0.5 * (ecg[:, 0] + ecg[:, 1])
    ecg[:, 4] = ecg[:, 0] - 0.5 * ecg[:, 1]
    ecg[:, 5] = ecg[:, 1] - 0.5 * ecg[:, 0]

    # Plot the ECG signal
    plt.figure(figsize=(20, 8))
    ecg_plot.plot_12(ecg.T * (2 / 1000), title=f"Deep Fake ECG {i}")

    # Save the plot as a JPEG file
    plt.savefig(f"ecg_plots_images/ecg_plot_{j}.jpg", dpi=300, bbox_inches="tight")

    # Show the plot
    plt.show()

    # Increment the index for the next plot
    j += 1

    # Close the plot to free memory
    plt.close()

print("All ECG plot images saved successfully.")
```
